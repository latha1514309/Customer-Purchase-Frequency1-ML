Customer Purchase Frequency Analysis using Machine Learning

## Internship Project

This project analyzes customer purchase behavior using Python and basic Machine Learning techniques.


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data = {
    'CustomerID': ['C001','C002','C001','C003','C002','C001'],
    'PurchaseDate': ['2023-01-05','2023-01-06','2023-01-15',
                     '2023-01-20','2023-01-25','2023-02-02'],
    'Amount': [500,300,700,450,600,800]
}

import pandas as pd

df = pd.DataFrame(data)
df['PurchaseDate'] = pd.to_datetime(df['PurchaseDate'])

df
```

```python
# Purchase count per customer
df['CustomerID'].value_counts()
```

```python
# Total amount spent by each customer
df.groupby('CustomerID')['Amount'].sum()

```

```python
# Bar graph for purchase frequency
df['CustomerID'].value_counts().plot(kind='bar')
plt.xlabel("Customer ID")
plt.ylabel("Number of Purchases")
plt.title("Customer Purchase Frequency")
plt.show()
```

```python
# Convert PurchaseDate to Day number for ML
df['Day'] = df['PurchaseDate'].dt.day

# Features and target
X = df[['Day']]
y = df['Amount']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Prediction
y_pred = model.predict(X_test)

# Check error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red')
plt.xlabel("Day")
plt.ylabel("Amount")
plt.title("Purchase Amount Prediction")
plt.show()
```

## Conclusion

This project analyzed customer purchase frequency using Python and a basic Machine Learning model (Linear Regression).
The results help identify frequent customers and predict purchase amount trends.
Such analysis can assist businesses in understanding customer behavior and improving sales strategies.
